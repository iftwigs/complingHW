```python
from bs4 import BeautifulSoup
import requests

page = requests.get('http://classic.newsru.com/main/04dec2018/')

parsed = BeautifulSoup(page.content, 'html.parser')
pagelinks = parsed.find_all('a', class_='explaincolumn')
pagelinks_clear = [i['href'] for i in pagelinks]

for i in pagelinks_clear:
    page_i = requests.get('http://classic.newsru.com' +i)
    parsed_i = BeautifulSoup(page_i.content, 'html.parser')
    all_rubbish = parsed_i.find_all('script')
    for m in all_rubbish:
        m.decompose()
    page_text = parsed_i.find('p')
    page_text_clear = page_text.get_text()
    with open (i.replace('/', '.')+'.txt', 'w', encoding='utf8') as openfile:
        openfile.write(page_text_clear)
```

```python
from bs4 import BeautifulSoup
import requests
import os

page = requests.get('http://classic.newsru.com/arch/')

parsed = BeautifulSoup(page.content, 'html.parser')
dayslinks = parsed.find_all('a', class_="arch-item-weekday") 

daylinks_clear = [i['href'] for i in dayslinks]
for link in daylinks_clear: 
    main_link = "http://classic.newsru.com" + link
    os.mkdir(link.replace('/', '.'))
    page_inner = requests.get(main_link)
    parsed_inner = BeautifulSoup(page_inner.content, 'html.parser')
    pagelinks = parsed_inner.find_all('a', class_='explaincolumn')
    for i in pagelinks:
        print(i['href'])
    pagelinks_clear = [i['href'] for i in pagelinks]
    for i in pagelinks_clear:
            page_i = requests.get('http://classic.newsru.com' +i)
            parsed_i = BeautifulSoup(page_i.content, 'html.parser')
            all_rubbish = parsed_i.find_all('script')
            for m in all_rubbish:
                m.decompose()
            page_text = parsed_i.find('p')
            page_text_clear = page_text.get_text()
            with open("/Users/macbookpro/Downloads" + link.replace('/', '.') + '\\' + i.replace('/', '.')+".txt", 'w', encoding='utf8') as openfile:
                openfile.write(page_text_clear)
```
